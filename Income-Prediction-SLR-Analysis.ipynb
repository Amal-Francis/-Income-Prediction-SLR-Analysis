```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
print("Number of instances",df.shape[0])
print("\nNumber of attributes",df.shape[1])
print("\nNumber of unique values\n",df.nunique())
print("\nNumber of null values\n",df.isnull().sum())
print("\nNumber of predicted values\n",df['per capita income (US$)'].value_counts())
```

```python
x = df['year'].values
y = df.drop('year', axis=1).values
```

```python
x = x.reshape(-1, 1)
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.score(x,y)
```

```python
# Predicting the capital income from 2020 to 2024
years = [2019,2020, 2021, 2023, 2024]
years = np.reshape(years, (-1,1))
```

```python
years = [1975, 1985, 1995, 2005, 2015]
years = np.reshape(years, (-1, 1))
plt.scatter(x,y, color='red')
plt.plot(years, reg.predict(years), color='green', lw=2)
```

```python
# Predicting the capital income of 2024
y_pred = reg.predict([[2024]])
y_pred
```
